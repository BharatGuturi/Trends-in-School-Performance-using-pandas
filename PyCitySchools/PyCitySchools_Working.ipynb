```python
#importing dependencies
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.to_csv('school_data_complete.csv', encoding='utf-8')
school_data_complete.head()
```

```python
#total number of schools
total_number_schools = school_data_complete['school_name'].nunique()
total_number_schools
```

```python
#total number of students
total_number_students = school_data_complete['student_name'].count()
total_number_students
```

```python
#total budget
school_names = school_data_complete['school_name'].unique()
school_budget = school_data_complete['budget'].unique()
school_budget_total = school_budget.sum()
print(school_budget_total)
```

```python
#per student budget
per_student_budget = school_budget_total/len(school_data_complete)
per_student_budget_round = print(round(per_student_budget,2))
```

```python
#average maths score
avg_maths_score = school_data_complete['maths_score'].sum()/len(school_data_complete)
avg_maths_score_round = round(avg_maths_score,2)
avg_maths_score_round
```

```python
#average reading score
avg_reading_score = school_data_complete['reading_score'].sum()/len(school_data_complete)
avg_reading_score_round = round(avg_reading_score,2)
avg_reading_score_round
```

```python
#% passing maths (the percentage of students who passed maths)
#assuming pass mark is 50
students_passed_maths = school_data_complete.loc[school_data_complete['maths_score']>= 50]
percent_students_passed_maths = round(len(students_passed_maths)*100/len(school_data_complete),2)
percent_students_passed_maths
```

```python
#% passing reading (the percentage of students who passed reading)
#assuming pass mark is 50
students_passed_reading = school_data_complete.loc[school_data_complete['reading_score']>= 50]
percent_students_passed_reading = round(len(students_passed_reading)*100/len(school_data_complete),2)
percent_students_passed_reading
```

```python
#% passing reading and maths(the percentage of students who passed reading and maths)
#assuming pass mark is 50
students_passed_reading_maths = school_data_complete.loc[(school_data_complete['reading_score']>= 50) & (school_data_complete['maths_score']>= 50)]
percent_students_passed_reading_maths = round(len(students_passed_reading_maths)*100/len(school_data_complete),2)
percent_students_passed_reading_maths
```

```python
output_data_LGA = {'Total Schools': [total_number_schools],
                   'Total Students': [total_number_students],
                   'Total Budget':  [school_budget_total],
                   'Average Maths Score': [avg_maths_score_round],
                   'Average Reading Score': [avg_reading_score_round],
                   '% Passing Maths': [percent_students_passed_maths],
                   '% Passing Reading': [percent_students_passed_reading],
                   '% Overall Passing': [percent_students_passed_reading_maths]
                  }
output_data_LGA = pd.DataFrame(output_data_LGA)
output_data_LGA
```

```python
output_data_LGA.loc[:,'Total Budget'] = output_data_LGA['Total Budget'].map('${:,d}'.format)
output_data_LGA.loc[:,'Total Students'] = output_data_LGA['Total Students'].map('{:,d}'.format)
output_data_LGA
```

```python
#creating summary table
school_Summary = pd.DataFrame({"School Type":school_Summary_groupby["type"].min(),
                               "Total Students":school_Summary_groupby.size(),
                               "Total School Budget":school_Summary_groupby["budget"].min(),
                               "Per Student Budget":round(school_Summary_groupby["budget"].min()/school_Summary_groupby.size(),2),
                               "Average Maths Score": round(school_Summary_groupby["maths_score"].sum()/school_Summary_groupby.size(),2),
                               "Average Reading Score": round(school_Summary_groupby["reading_score"].sum()/school_Summary_groupby.size(),2),
                               "% Passing Maths": round(students_passed_maths_groupby['maths_score'].size()*100/school_Summary_groupby.size(),2),
                               "% Passing Reading": round(students_passed_reading_groupby['reading_score'].size()*100/school_Summary_groupby.size(),2),
                               "% Overall Passing": round(students_passed_reading_maths_groupby['maths_score'].size()*100/school_Summary_groupby.size(),2)
                               })
school_Summary
school_spending_initial = school_Summary.copy()
school_size_initial = school_Summary.copy()
school_type_initial = school_Summary.copy()
```

```python
#Currency formatting
school_Summary.loc[:,'Total School Budget'] = school_Summary['Total School Budget'].map('${:,.2f}'.format)
school_Summary.loc[:,'Per Student Budget'] = school_Summary['Per Student Budget'].map('${:,.2f}'.format)
school_Summary
```
